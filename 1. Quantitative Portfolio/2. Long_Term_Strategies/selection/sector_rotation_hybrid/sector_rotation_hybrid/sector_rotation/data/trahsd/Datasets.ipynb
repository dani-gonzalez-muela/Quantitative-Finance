```python
# Check the actual underlying source for each table
tables = [
    ('comp_na_daily_all', 'fundq'),
    ('comp_na_daily_all', 'secd'),
    ('comp_global_daily', 'g_fundq'),
    ('comp_global_daily', 'g_secd'),
    ('wrdsapps_finratio', 'firm_ratio'),
    ('contrib_global_factor', 'global_factor'),
    ('crsp_a_ccm', 'ccmxpf_lnkhist'),
]

for schema, table in tables:
    info = db.raw_sql(f"""
        SELECT table_schema, table_name, table_type 
        FROM information_schema.tables 
        WHERE table_schema = '{schema}' AND table_name = '{table}'
    """)
    print(f"{schema}.{table} -> {info['table_type'].values[0]}")
```

```python
# 1. NA Fundamentals Quarterly
print("=== 1. NA Fundamentals Quarterly ===")

fundq = db.raw_sql("""
    SELECT * FROM comp_na_daily_all.fundq
    WHERE consol = 'C'
      AND indfmt IN ('INDL', 'FS')
      AND datafmt = 'STD'
      AND popsrc IN ('D', 'I')
      AND curcdq = 'USD'
    LIMIT 5
""")
print(fundq)
```

```python
# 2. NA Security Daily
print("\n=== 2. NA Security Daily ===")

secd = db.raw_sql("""
    SELECT * FROM comp_na_daily_all.secd
    LIMIT 5
""")
print(secd)
```

```python
# 3. Global Fundamentals Quarterly
print("\n=== 3. Global Fundamentals Quarterly ===")

g_fundq = db.raw_sql("""
    SELECT * FROM comp_global_daily.g_fundq
    WHERE consol = 'C'
      AND indfmt IN ('INDL', 'FS')
      AND datafmt = 'HIST_STD'
    LIMIT 5
""")
print(g_fundq)
```
